```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
import plotly.plotly as py
```

```python
df = pd.read_csv('h1b_kaggle.csv')
num_of_rows = float(len(df))
```

```python
df=df.dropna()
number_of_complete_rows = float(len(df))

print'Number of Rows', num_of_rows
print'Number of Complete Rows', number_of_complete_rows
print'Number of Rows loss', num_of_rows - number_of_complete_rows
print'Precentage lost', float(number_of_complete_rows/ num_of_rows)
```

#### We can look at the Certified and Denied since that is the bulk of the data.

```python
df.EMPLOYER_NAME.value_counts().sort_values(ascending=False).head(10)
```

# Top ten Firms are IT related. Interesting!

```python
# H1-B Visa by Years
sns.set_context("notebook",font_scale=1.0)
plt.figure(figsize=(10,3))
plt.title('PETITIONS DISTRIBUTION BY YEAR')
sns.countplot(df['YEAR'])
```

## We only care about either being approve or denied

```python
# H1-B Visa by Status
sns.set_context("notebook",font_scale=1.0)
plt.figure(figsize=(10,3))
df4=df[df['CASE_STATUS'].isin(['CERTIFIED','DENIED'])]
df4.groupby(['CASE_STATUS']).size().plot(kind='barh')
```

```python
# Top 30 Job Titles
sns.set_context("notebook",font_scale=1.0)
plt.figure(figsize=(10,10))
df.JOB_TITLE.value_counts().sort_values(ascending=False).head(30).plot(kind='barh')
```
